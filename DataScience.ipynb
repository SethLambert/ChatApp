```python
#difference between 10th and 90th percent, lower difference is lower risk
clean_df['Mid-Career 90th Percentile Salary'].subtract(clean_df['Mid-Career 10th Percentile Salary'])
```

```python
spread_col = clean_df['Mid-Career 90th Percentile Salary'] - clean_df['Mid-Career 10th Percentile Salary']
clean_df.insert(1, 'Spread', spread_col)
clean_df.head()
```

```python
low_risk = clean_df.sort_values('Spread')
low_risk[['Undergraduate Major', 'Spread']].head()
```

```python
low_risk
```

```python
#top five highest mid-career
highest_90_at_10 = clean_df.sort_values('Mid-Career 90th Percentile Salary', ascending=False)
highest_90_at_10[['Undergraduate Major', 'Mid-Career 90th Percentile Salary']].head()
```

```python
#highest risk/spread
high_risk = clean_df.sort_values('Spread', ascending = False)
high_risk[['Undergraduate Major', 'Spread']].head()
```

```python
#highest risk/spread
highest_median_at_10 = clean_df.sort_values('Mid-Career Median Salary', ascending = False)
highest_median_at_10[['Undergraduate Major', 'Mid-Career Median Salary']].head()
```
